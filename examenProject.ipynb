```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv") #reading csv file
```

```python
df.head()
```

```python
counts = df["type"].value_counts()
plt.bar(counts.index, counts.values) #matplotlib Bar default chart
plt.show()
```

```python
df["type"].value_counts().plot.bar() #Pandas default plot bar 
plt.show()
```

```python
sns.countplot(x="type", data=df) #Seaborn Barchart 
plt.show()
```

```python
df = df[df['type'] == 'Movie']
```

```python
df['duration'] = df['duration'].map(lambda x: x.rstrip('min')).astype(int)
```

```python
df_new = df
df_new['duration'].hist(bins=100)
```

```python
from collections import Counter
labels, communSizes = zip(*Counter(df['country'].values).most_common(5))
genres, genreSizes = zip(*Counter(df['country'].values).items())
labels = list(labels)
communSizes = list(communSizes)
labels.append('other')
communSizes.append(sum(list(genreSizes))-sum(communSizes))
fig, ax = plt.subplots()
ax.pie(communSizes, labels=labels, autopct='%1.1f%%')
ax.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
ax.set_title('genres')

plt.show()
```
